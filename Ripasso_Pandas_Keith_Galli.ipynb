Mettiamo che vogliamo classificare le persone tra persone basse medie e alte o qualcosa del genere.

```python
bios["height_category"] = bios["height_cm"].apply(lambda x: "Short" if x <165 else ("Average" if x < 185 else "Tall"))
```

Con la funzione lamba noi andiamo a scrivere una funzione senza usare def, in questo caso abbiamo scritto una funzione tutta sulla riga.

```python
bios.head()

#Abbiamo quindi in base a determinate condizioni creato la colonna con Tall/Average/Short
```

```python
def categorize_athlete(row):
  if row["height_cm"] < 175 and row["weight_kg"] < 70:
    return "Lightweight"
  elif row["height_cm"] <185 or row["weight_kg"] <= 80:
    return "Middleweight"

  else:
    return "Heavyweight"

bios["Category"] = bios.apply(categorize_athlete, axis=1)
```

Qui abbiamo creato una funzione che va a categorizzare, poi abbiamo usato il metodo apply per applicare questa funzione lungo le righe (axis=1) e abbiamo assegnato tutti questi valori ad una colonna [" Category “]

# Handling Null Values

```python
coffe.loc[[0,1],"Units Sold"] = np.nan
```

```python
coffe.head(4)
```

Siamo andati a inserire dei valori NaN, e questi li possiamo controllare tramite info.

```python
coffe.info()
```

Vediamo che nella colonna Units Sold abbiamo 12 non-null quindi 2 sono nulli. Quindi possiamo vederne la presenza quando usiamo il metodo info.

```python
coffe.isna().sum()

#con questo comando andiamo a determinare quanti na abbiamo
```

Quando abbiamo delle posizioni con na possiamo andare a fillarle.

```python
coffe = coffe.fillna(coffe["Units Sold"].mean())

#qui andiamo a sostituire gli na con i valori medi
```
