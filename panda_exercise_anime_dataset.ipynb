Here below we gonna open the .csv file, see the main structure of the DataFrame and get some general info about the dataset.

```python
anime = pd.read_csv("anime.csv")
anime.info()
anime.head(5)
print("\n")
print(f"The dataset has {anime.shape[0]} lines and {anime.shape[1]} columns")

```

```python
anime.describe()

# general info about the Dataframe, the only one ueful is the rating colum
```

```python
anime["genre"].unique()

# the categorical values in the genre column
```

```python
rating = pd.read_csv("rating.csv")
rating.info()
rating.head(5)

# exploration about the "rating" csv file
```

```python
anime["genre"].value_counts()
```

```python
grouped_anime_rating = anime.groupby("type")["rating"].mean()
grouped_anime_rating

# It's grouping base on the "type" and then it's giving the mean for each "type" category
```

```python
top10_anime_members = anime.sort_values("members",ascending=False ).head(10)
top10_anime_members

#These are the top 10 anime by number of members
```

```python
# Filtration animes with major 8 rating value and with more than 10.000 members

anime_filtrated = anime[(anime["rating"] > 8) & (anime["members"] >10000)]
anime_filtrated["genre"] = anime_filtrated["genre"].str.split(',')

anime_exploded = anime_filtrated.explode("genre")

genre_counts = anime_exploded.groupby("genre")["genre"].size()


print(genre_counts.sort_values(ascending=False))

# Among the most called genre we have Action, Shounen and Comedy
```
