```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creazione delle date giornaliere per l'anno 2022
date_range = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')

# Generazione di dati casuali per le vendite e i ritorni
np.random.seed(0)  # Imposta il seed per la riproducibilità
vendite = np.random.randint(50, 200, size=len(date_range))  # Numeri casuali tra 50 e 200
ritorni = np.random.randint(0, 50, size=len(date_range))    # Numeri casuali tra 0 e 50

# Assicurarsi che le vendite siano sempre maggiori o uguali ai ritorni
ritorni = np.minimum(vendite, ritorni)

# Creazione del DataFrame
df_vendite = pd.DataFrame({
    'Data': date_range,
    'Vendite': vendite,
    'Ritorni': ritorni
})

# Visualizzazione delle prime righe del DataFrame
print(df_vendite.head())

# 365 row and 3 columns

df_vendite.shape

```

```python
# Calcola il totale delle vendite e dei ritorni per l'anno

vendite_year = df_vendite["Vendite"].sum()
ritorni_year = df_vendite["Ritorni"].sum()

print(f"Le vendite annuali sono: {vendite_year} e i ritorni: {ritorni_year}")
```

```python
df_vendite.info()
```

# <mark> Aggregazione dei dati

```python

# Aggiungiamo colonna per il mese e dato che "data" è di tipo datetime

df_vendite["Mese"] = df_vendite["Data"].dt.to_period("M")

# Abbiamo aggregato per i vari Mesi e abbiamo eseguito per ogni mese la somma
# delle vendite e dei ritorni

vendite_mensili = df_vendite.groupby("Mese")["Vendite"].sum()
ritorni_mensili = df_vendite.groupby("Mese")["Ritorni"].sum()

# Visualizza i risultati
print("Vendite Mensili:")
print(vendite_mensili)

print("\nRitorni Mensili:")
print(ritorni_mensili)
```

```python
# Vendite nette, creiamo una nuova colonna vendite_nette
# E in quella colonna metto al differenza tra vendite e ritorni

df_vendite["Vendite nette"] = df_vendite["Vendite"] - df_vendite["Ritorni"]
```

```python
vendite_mensili.head(4)
```

```python

# Configura il grafico
fig, ax = plt.subplots(figsize=(12, 6))

# Indici per il posizionamento delle barre quindi rappresenta gli indici dei mesi
# creati grazie
ind = range(len(vendite_mensili))

# Larghezza delle barre
width = 0.4

# Crea le barre per le vendite e i ritorni

# Ogni ax.bar crea un insieme di barre che hanno l'x e poi l'y

# Le barre di ritorni_mensili hanno una posizione lungo x leggermente spostata
# l'indici + 0.4 per ogni indice in ind, alpha 0.7 indica che sono leggermente trasparenti

ax.bar(ind, vendite_mensili, width, label='Vendite Mensili')
ax.bar([i + width for i in ind], ritorni_mensili, width, label='Ritorni Mensili', alpha=0.7)

# Aggiungi etichette, titolo e legenda
ax.set_xlabel('Mese')
ax.set_ylabel('Totale')
ax.set_title('Vendite e Ritorni Mensili nel 2022')
ax.set_xticks([i + width / 2 for i in ind])
# Impostazione delel date lungo l'asse x
ax.set_xticklabels(vendite_mensili.index.astype(str), rotation=45)
# Inserimento della legenda
ax.legend()

# Mostra il grafico
plt.tight_layout()
plt.show()
```
